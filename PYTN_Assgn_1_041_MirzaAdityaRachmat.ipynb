## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

How has major crime been improved in Sutton between 2013-2016?

```python
# Write your function below
by_2016 = by_year.loc[2016]

Sutton_2016 = by_2016["borough"] == "Sutton"

Sutton_2016_major = by_2016[(Sutton_2016)]["Major"].value_counts()



by_2015 = by_year.loc[2015]

Sutton_2015 = by_2015["borough"] == "Sutton"

Sutton_2015_major = by_2015[(Sutton_2015)]["Major"].value_counts()
S_2015_list = list(Sutton_2015_major)



by_2014 = by_year.loc[2014]

Sutton_2014 = by_2014["borough"] == "Sutton"

Sutton_2014_major = by_2014[(Sutton_2014)]["Major"].value_counts()
S_2014_list = list(Sutton_2014_major)



by_2013 = by_year.loc[2013]

Sutton_2013 = by_2013["borough"] == "Sutton"

Sutton_2013_major = by_2013[(Sutton_2013)]["Major"].value_counts()
S_2013_list = list(Sutton_2013_major)





Sutton = pd.DataFrame(Sutton_2016_major)
Sutton.rename(columns={"Major":"2016"}, inplace=True)
Sutton.insert(1,"2015",S_2015_list,True)
Sutton.insert(1,"2014",S_2014_list,True)
Sutton.insert(1,"2013",S_2013_list,True)
Sutton.sort_index(axis=1, inplace=True)
Sutton_ = Sutton.transpose()



# Graded-Funtion Begin (~1 Lines)
Sutton_.plot(kind="area", figsize=(15,10))
# Graded-Funtion End

plt.title('Major Crime in Sutton between 2013-2016') # add a title to the area plot
plt.ylabel('Number of Crime Recorded') # add y-label
plt.xlabel('Years') # add x-label
plt.legend(loc="lower left")

plt.show()
```

**Insight:**

For over 4 years, violence against the person has been the highest major crime recorded in the borough of Sutton. But in 2016, theft and handling came in close for the highest. Robbery has been the lowest since 2013. Most of the crimes were increased, but violence against the person has been steady.

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

How consistent did major crimes happened in Sutton in 2013-2016?

```python
# Write your function below

# Graded-Funtion Begin (~2 Lines)
Sutton_.plot(kind="hist", figsize=(15,10),
             alpha=.7,
            )
# Graded-Funtion End

plt.title('Histogram of Major Crime Recorded in Sutton in 2013-2016') # add a title to the histogram
plt.ylabel('Frequency between 2013-2016') # add y-label
plt.xlabel('Number of Crimes Recorded') # add x-label
plt.legend(loc="lower left")

plt.show()
```

**Insight:**

With histogram, we can see the frequency of these crime in 2013-2016. Although robbery has the lowest crime recorded, it consistently contributed in crime recorded in 4 years. Same goes with violence, but with higher number of crimes recorded.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

How many theft crimes were recorded from 2008 to 2016 in each borough in London?

```python
# Write your function below
drugs = by_year["Major"] == "Drugs"
drugs_df = by_year[(drugs)]["borough"].value_counts()
df_drugs = pd.DataFrame(drugs_df)
df_drugs.reset_index(inplace=True)
df_drugs.columns = ["borough", "total"]

# Graded-Funtion Begin (~1 Lines)
drugs_df.plot(kind='barh', figsize=(10,6))
# Graded-Funtion End

plt.xlabel('Borough') # add to x-label to the plot
plt.ylabel('Crime counts') # add y-label to the plot
plt.title('Theft Counts in London 2008-2016') # add title to the plot

plt.show()
```

**Insight:**

With bar chart, we can see how theft crimes were recorded in between 2008 and 2016. Southwark has the most crimes recorded and Brent came in second. City of london has the lowest theft recorded in these years.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Which major crime occured the most in Sutton area in 2020?

```python
# Write your function below

# Graded-Funtion Begin (~8 Lines)
Sutton_2016_major.plot(kind='pie',
                      figsize=(14, 6),
                      autopct='%1.1f%%',
                      startangle=90,
                      shadow=True,         
                      labels=None,
                      pctdistance=1.10
)
# Graded-Funtion End

# scale the title up by 12% to match pctdistance
plt.title('Major Crime in Sutton in 2016', y=1.12)

plt.axis('equal')

# add legend
plt.legend(labels=Sutton_2016_major.index, loc='upper left')

plt.show()
```

**Insight:**

According to the chart, Violence against the Person is the most occured crime in Sutton in 2016. Followed by Theft and Handling, while Robbery was the least occured major crime in 2016.

Pie chart is used to see the percentage of crimes that occured in Sutton in 2016.

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

What is the statistics of violence against the person in London in 2016?

```python
# Write your function below
Violence_2016 = by_2016["Major"] == "Violence Against the Person"

Violence_2016_bor = by_2016[(Violence_2016)]["borough"].value_counts()

# Graded-Funtion Begin (~1 Lines)
Violence_2016_bor.plot(kind="box", figsize=(8, 6))
# Graded-Funtion End

plt.title('Statistics of Violence Against the Person in 2016')
plt.ylabel('Total Violence Against the Person Crime Count per Borough in 2016')

plt.show()
```

**Insight:**

The range of crime count of violence against the person in 2016 is between 1747 to 5406 which are in borough of Kingston upon Thames and Lambeth respectively. City of London is the outlier with only 24 crimes recorded in 2016. the chart is skewed to the right although it's not significant.

Box chart is used to show the skewness of the data and the only one outlier can be shown clearly.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

How major crimes related to months in a year?

```python
# Write your function below
month_sutton = by_year[(by_year["borough"] == "Sutton")]["month"].value_counts()
df_month_sutton = pd.DataFrame(month_sutton)
df_month_sutton.reset_index(inplace=True)
df_month_sutton.columns = ["month", "total"]
df_month_sutton['month'] = df_month_sutton["month"].astype(int)
df_month_sutton.head()

# Graded-Funtion Begin (~1 Lines)

df_month_sutton.plot(
        kind='scatter', 
        x='month', 
        y='total', 
        figsize=(6,5), 
        color='darkblue'
)
# Graded-Funtion End

plt.title('Major Crime in Sutton by months')
plt.xlabel('Month')
plt.ylabel('Major Crime Count')

plt.show()
```

**Insight:**

Scatter plot is used to see each major crime counts in each month. December is the least major crime count in a year ranging between 2008-2016. The highest major crime count is in June. This could occured because of the weather and season. Could also because of Holiday season in December.

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
df_drugs_word = df_drugs.set_index('borough')
total_drugs = df_drugs_word['total'].sum()

max_words = 90
word_string = ''

for drug in df_drugs_word.index.values:
    if len(drug.split(' ')) == 1:
        repeat_num_times = int(df_drugs_word.loc[drug, 'total']/float(total_drugs)*max_words)
        word_string = word_string + ((drug + ' ') * repeat_num_times)

# instantiate a word cloud object
your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
your_wordcloud.generate(word_string)
```

Which borough did have the most drugs-related crime?

```python
# Write your function below
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(12)

# Graded-Funtion Begin (~1 Lines)
plt.imshow(your_wordcloud, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```

With wordcloud, we can see that Southwark and Brent are the most borough drugs-related crime. City of London with the lowest number doesn't appear in it because it is an outlier. 

## Folium

Folium is a powerful Python library that helps you create several types of Leaflet maps. The fact that the Folium results are interactive makes this library very useful for dashboard building.

From the official Folium documentation page:

> Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in on a Leaflet map via Folium.

> Folium makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

> The library has a number of built-in tilesets from OpenStreetMap, Mapbox, and Stamen, and supports custom tilesets with Mapbox or Cloudmade API keys. Folium supports both GeoJSON and TopoJSON overlays, as well as the binding of data to those overlays to create choropleth maps with color-brewer color schemes.

```python
# define the world map
world_map = folium.Map(location=[0,0], zoom_start=2)

# display world map
world_map
```

```python
latitude = 51.509865
longitude = -0.118092

# define the world map centered around London with a low zoom level 8
london_map = folium.Map(location=[latitude, longitude], zoom_start=8)

# Write your function below
sutton = folium.CircleMarker(
        [51.3696371, -0.1987839],
        radius=8,
        color="yellow",
        fill=True,
        fill_color="blue",
        fill_opacity=0.6
)
# Graded-Funtion Begin (~1 Lines)
london_map.add_child(sutton)
# Graded-Funtion End
```
